```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
death = pd.read_csv('COVID DEATHS.csv', na_values=':', usecols = ["location","date","population","total_cases","new_cases","total_deaths","new_deaths","total_deaths","new_deaths","reproduction_rate"])

display(death)
```

```python
philippines_death = death[death["location"] == "Philippines"]

philippines_death = philippines_death[["date", "new_deaths", "new_cases","reproduction_rate"]]

display(philippines_death)
```

```python
descriptive_stats = philippines_death[["new_deaths", "new_cases", "reproduction_rate"]].describe()
display(descriptive_stats)
```

```python
philippines_death['date'] = pd.to_datetime(philippines_death['date'])

plt.figure(figsize=(40, 10))
plt.plot(philippines_death['date'], philippines_death['new_deaths'], label='Daily mortality rate')
plt.title('Philippine daily COVID-19 mortality count')
plt.xlabel('Date')
plt.ylabel('Mortality rate')
plt.legend()
plt.show()

plt.figure(figsize=(40, 10))
plt.plot(philippines_death['date'], philippines_death['new_cases'], label='New Cases', color='orange')
plt.title('Philippine Daily Case Count')
plt.xlabel('Date')
plt.ylabel('New Cases')
plt.legend()
plt.show()
```

```python
philippines_death['death_to_case_ratio'] = philippines_death['new_deaths'] / philippines_death['new_cases']

print(philippines_death[['date', 'death_to_case_ratio']])
```

```python
plt.figure(figsize=(40, 10))
plt.plot(philippines_death['date'], philippines_death['death_to_case_ratio'], label='Death-to-Case Ratio', color='purple')

plt.title('Death-to-Case Ratio Over Time in the Philippines')
plt.xlabel('Date')
plt.ylabel('Ratio')
plt.legend()

plt.show()
```

```python
vaccine = pd.read_csv('COVID_VACCINATIONS.csv', na_values=':', usecols = ["location","date","total_tests","new_tests","total_vaccinations","people_vaccinated","people_fully_vaccinated","total_boosters","new_vaccinations","stringency_index"])

display(vaccine)
```

```python
philippines_vaccine = vaccine[vaccine["location"] == "Philippines"]

philippines_vaccine = philippines_vaccine[["location","date","total_tests","new_tests","total_vaccinations","people_vaccinated","people_fully_vaccinated","total_boosters","new_vaccinations","stringency_index"]]

display(philippines_vaccine)
```

```python
descriptive_stats_vaccines = philippines_vaccine[["location","date","total_tests","new_tests","total_vaccinations","people_vaccinated","people_fully_vaccinated","total_boosters","new_vaccinations","stringency_index"]].describe()
display(descriptive_stats_vaccines)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer
from sklearn.metrics import mean_squared_error

# Assuming 'philippines_death' and 'philippines_vaccine' are your original dataframes

# Convert 'date' column in 'philippines_vaccine' to datetime64[ns]
philippines_vaccine['date'] = pd.to_datetime(philippines_vaccine['date'])

# Modify the merge based on your actual column names
merged_data = pd.merge(philippines_death, philippines_vaccine[['date', 'people_vaccinated', 'people_fully_vaccinated', 
                                                                'total_boosters', 'new_vaccinations', 
                                                                'stringency_index']], on="date", how="inner")

# Update features and target
features = ['people_vaccinated', 'people_fully_vaccinated', 'total_boosters', 
            'new_vaccinations', 'stringency_index']

target = "new_deaths"

# Impute missing values in features
imputer_features = SimpleImputer(strategy='mean')
merged_data_imputed = pd.DataFrame(imputer_features.fit_transform(merged_data[features]), columns=features)

# Impute missing values in the target variable
imputer_target = SimpleImputer(strategy='mean')
merged_data[target] = imputer_target.fit_transform(merged_data[[target]])

# Create a heatmap for correlations
plt.figure(figsize=(12, 8))
sns.heatmap(merged_data[['new_deaths', 'people_vaccinated', 'people_fully_vaccinated', 
                         'total_boosters', 'new_vaccinations', 
                         'stringency_index']].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()

# Perform linear regression and create scatter plots for each feature
for feature in features:
    X_train, X_test, y_train, y_test = train_test_split(merged_data_imputed, merged_data[target], test_size=0.2, random_state=50)

    if len(X_train) == 0:
        print(f"Adjust parameters for {feature}, train set is empty.")
    else:
        model = LinearRegression()
        model.fit(X_train, y_train)

        predictions = model.predict(X_test)

        mse = mean_squared_error(y_test, predictions)
        print(f"\nLinear Regression Results for {feature}:")
        print(f"Mean Squared Error: {mse}")

        plt.figure(figsize=(10, 6))
        plt.scatter(X_test[feature].values, y_test.values, label='Actual')
        plt.scatter(X_test[feature].values, predictions, label='Predicted')
        plt.title(f'Actual vs. Predicted New Deaths for {feature}')
        plt.xlabel(feature)
        plt.ylabel('New Deaths')
        plt.legend()
        plt.show()
```

```python
# Create a linear regression model
model = LinearRegression()

# Train the model on the training set
model.fit(X_train, y_train)

# Make predictions on the testing set
predictions = model.predict(X_test)

# Calculate and print the Mean Squared Error
mse = mean_squared_error(y_test, predictions)
print(f"Mean Squared Error: {mse}")

# Plot the scatter plot with the linear regression line
plt.figure(figsize=(10, 6))
plt.scatter(X_test.values, y_test.values, label='Actual')
plt.plot(X_test.values, predictions, color='red', label='Linear Regression Line')
plt.title('New Deaths vs. People Vaccinated')
plt.xlabel('People Vaccinated')
plt.ylabel('New Deaths')
plt.legend()
plt.show()
```
